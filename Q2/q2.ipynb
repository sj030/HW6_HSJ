```python
# 시행 결과를 각각 csv 파일에 저장
import random
import csv

f = open('onehundred.csv', 'w', newline='')
writer = csv.writer(f)

for i in range(100):
    onehundred = [random.randint(1,6)]
    writer.writerow(onehundred)
f.close()
```

```python
f = open('onethousand.csv', 'w', newline='')
writer = csv.writer(f)

for i in range(1000):
    onethousand = [random.randint(1,6)]
    writer.writerow(onethousand)
f.close()
```

```python
f = open('tenthousand.csv', 'w', newline='')
writer = csv.writer(f)

for i in range(10000):
    tenthousand = [random.randint(1,6)]
    writer.writerow(tenthousand)
f.close()
```

```python
f = open('onehundredthousand.csv', 'w', newline='')
writer = csv.writer(f)

for i in range(100000):
    onehundredthousand = [random.randint(1,6)]
    writer.writerow(onehundredthousand)
f.close()
```

```python
# 시행별 데이터를 읽어 와, 그대로 저장해 놓을 리스트 > 히스토그램 제작에 사용
result = [] #전체
resultOH = [] #OneHundred
resultOT = [] #OneThousand
resultTT = [] #TenThousand
resultOHT = [] #OneHundredThousand

#시뮬레이션 결과를 분석하기 위한 리스트
#시행에서 각 결과별로(1,2,3,4,5,6) 몇 회 발생했는지 저장.
countOH = [0,0,0,0,0,0]
countOT = [0,0,0,0,0,0]
countTT = [0,0,0,0,0,0]
countOHT = [0,0,0,0,0,0]
```

```python
f = open("onehundred.csv", "r")
reader = csv.reader(f)

for row in reader:
    num = int(row[0])
    result.append(num)
    resultOH.append(num)
    if num == 1:
        countOH[0] += 1
    elif num == 2:
        countOH[1] += 1
    elif num == 3:
        countOH[2] += 1
    elif num == 4:
        countOH[3] += 1
    elif num == 5:
        countOH[4] += 1
    elif num == 6:
        countOH[5] += 1
f.close()
```

```python
f = open("onethousand.csv", "r")
reader = csv.reader(f)

for row in reader:
    num = int(row[0])
    result.append(num)
    resultOT.append(num)
    if num == 1:
        countOT[0] += 1
    elif num == 2:
        countOT[1] += 1
    elif num == 3:
        countOT[2] += 1
    elif num == 4:
        countOT[3] += 1
    elif num == 5:
        countOT[4] += 1
    elif num == 6:
        countOT[5] += 1
    
f.close()
```

```python
f = open("tenthousand.csv", "r")
reader = csv.reader(f)

for row in reader:
    num = int(row[0])
    result.append(num)
    resultTT.append(num)
    if num == 1:
        countTT[0] += 1
    elif num == 2:
        countTT[1] += 1
    elif num == 3:
        countTT[2] += 1
    elif num == 4:
        countTT[3] += 1
    elif num == 5:
        countTT[4] += 1
    elif num == 6:
        countTT[5] += 1
f.close()
```

```python
f = open("onehundredthousand.csv", "r")
reader = csv.reader(f)

for row in reader:
    num = int(row[0])
    result.append(num)
    resultOHT.append(num)
    if num == 1:
        countOHT[0] += 1
    elif num == 2:
        countOHT[1] += 1
    elif num == 3:
        countOHT[2] += 1
    elif num == 4:
        countOHT[3] += 1
    elif num == 5:
        countOHT[4] += 1
    elif num == 6:
        countOHT[5] += 1
f.close()
```

```python
#onehundredthousand 히스토그램
plt.hist(resultOHT, bins=12)
plt.title("one-hundred-thousand")
plt.xlabel("dice result")
plt.ylabel("number of occurrences")
plt.show()
```

```python
# 표준 편차를 통해 분산 정도 알아보기 
# 각 1,2,...6의 발생 횟수가 고르게 나오는지 알아보자.

#1. 평균 구하기 (각 주사위의 값이 발생하는 횟수의 평균)
meanOH = sum(countOH) / 100
meanOT = sum(countOT) / 1000
meanTT = sum(countTT) / 10000
meanOHT = sum(countOHT) / 100000

#2. 분산 구하기
#onehundred
varianceOH = 0
for i in range(6):
    deviation = countOH[i] - meanOH #편차 
    varianceOH += deviation ** 2 #편차 제곱의 합
varianceOH /= 100

#onethousand
varianceOT = 0
for i in range(6):
    deviation = countOT[i] - meanOT
    varianceOT += deviation ** 2
varianceOT /= 1000

#tenthousand
varianceTT = 0
for i in range(6):
    deviation = countTT[i] - meanTT
    varianceTT += deviation ** 2
varianceTT /= 10000

#onehundredthousand
varianceOHT = 0
for i in range(6):
    deviation = countOHT[i] - meanOHT
    varianceOHT += deviation ** 2
varianceOHT /= 100000

#3. 표준 편차 구하기
varianceOH = varianceOH**(1/2)
varianceOT = varianceOT**(1/2)
varianceTT = varianceTT**(1/2)
varianceOHT = varianceOHT**(1/2)
```

```python
plt.title('Standard deviation')
xdata = ["100","1000","10000","100000"]
ydata = [varianceOH, varianceOT, varianceTT, varianceOHT]
plt.scatter(xdata, ydata)
```
