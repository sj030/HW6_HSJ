```python
import csv
f = open('2303.csv', 'r', encoding='ANSI')
data = csv.reader(f)
header1 = next(data)
header2 = next(data)
print(header1)

stationName = []
allRide = [] #7:00 ~ 8:59 모든 승차인 수
allQuit = [] #7:00 ~ 8:59 모든 하차인 수
allRideNQuit = [] #7:00 ~ 8:59 모든 승하차인 수

for row in data:
    stationName.append(row[3])
    ride = int(row[-2]) + int(row[-4])
    quit = int(row[-1]) + int(row[-3])
    allRide.append(ride)
    allQuit.append(quit)
    allRideNQuit.append(int(ride + quit))

f.close()
```

```python
# 출근시간 최대 승차역 30개 추출
maxRide = []
maxRideStationName = []
for i in range(30):
    maxRide.append(max(allRide))
    #값 제거
    idx = allRide.index(max(allRide))
    allRide[idx] = 0
    maxRideStationName.append(stationName[idx])
    
print(maxRide) #정렬되어 있음
```

```python
# 출근시간 최대 하차역 30개 추출
maxQuit = []
maxQuitStationName = []
for i in range(30):
    maxQuit.append(max(allQuit))
    idx = allQuit.index(max(allQuit))
    allQuit[idx] = 0
    maxQuitStationName.append(stationName[idx])
    
print(maxQuit)
```

```python
# 출근시간 최대 승하차역 30개 추출
maxRideNQuit = []
maxRideNQuitStationName = []
for i in range(30):
    maxRideNQuit.append(max(allRideNQuit))
    idx = allRideNQuit.index(max(allRideNQuit))
    allRideNQuit[idx] = 0
    maxRideNQuitStationName.append(stationName[idx])
    
print(maxRideNQuit)
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc # 폰트 세팅을 위한 모듈 추가
font_path = "C:/Windows/Fonts/malgun.ttf" # 사용할 폰트명 경로 삽입
font = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font)
```

```python
# 출근시간 최대 승차역 30개 시각화
plt.figure(figsize=(45, 5)) #그래프 크기 조절
plt.bar(maxRideStationName, maxRide)
plt.title("Max - ride")
plt.xlabel("name of station")
plt.ylabel("number of passengers")
plt.show()
```

```python
# 출근시간 최대 하차역 30개 시각화
plt.figure(figsize=(45, 5)) #그래프 크기 조절
plt.bar(maxQuitStationName, maxQuit)
plt.title("Max - quit")
plt.xlabel("name of station")
plt.ylabel("number of passengers")
plt.show()
```

```python
# 출근시간 최대 승하차역 30개 시각화
plt.figure(figsize=(45, 5)) #그래프 크기 조절
plt.bar(maxRideNQuitStationName, maxRideNQuit)
plt.title("Max - ride and quit")
plt.xlabel("name of station")
plt.ylabel("number of passengers")
plt.show()
```
